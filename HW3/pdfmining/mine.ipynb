```python

# a very rudimentary script for parsing data.

failures = []

def parse_line(line: str): 
    j = 0
    while line[j].isdigit(): j += 1
    incident_id = int(line[:j])

    i = j
    while not line[j].isdigit(): j += 1
    offense = line[i:j].strip()

    i = j
    while line[j-4:j] != '2024': j += 1
    date = line[i:j]

    while not line[j].isdigit(): j += 1 

    i = j 
    while line[j].isdigit(): j += 1
    street_number = int(line[i:j])

    i = j
    while not line[j].isdigit(): j += 1
    street = line[i:j].strip()

    i = j
    while line[j].isdigit(): j += 1
    hour = line[i:j]

    while line[j].isspace(): j += 1

    dow = line[j:j+2]

    j += 2
    district = int(line[j:])
    
    print(incident_id, offense, date, street_number, street, hour, dow, district)
        

for i, page in enumerate(reader.pages): 
    text = page.extract_text().split('\n')
    if "MAJOR CRIMES BETWEEN" in text[0]:
        # print(i, text[0])
        # print(text[1])
        for line in text:
            try:
                parse_line(line)
            except:
                failures.append(line)
                # print(line)
```
